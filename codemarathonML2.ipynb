```
import pandas as pd
import numpy as np
```

```
data=pd.read_csv('train.csv')
```

```
data.head()
```

```
dummy=np.array(data)
np.random.shuffle(dummy)
```

```
from keras.models import Sequential
from keras.layers import *
```

```
m1=Sequential()
```

```
m1.add(InputLayer(input_shape=(64,)))
m1.add(BatchNormalization(axis=-1))
m1.add(Dense(128,activation='relu'))
m1.add(Dropout(0.2))
m1.add(BatchNormalization(axis=-1))
m1.add(Dense(64, activation='relu'))
m1.add(Dropout(0.4))
m1.add(BatchNormalization(axis=-1))
m1.add(Dense(16, activation='relu'))
m1.add(Dropout(0.5))
m1.add(BatchNormalization(axis=-1))
m1.add(Dense(16, activation='relu'))
m1.add(Dropout(0.6))
m1.add(BatchNormalization(axis=-1))
m1.add(Dense(2, activation='softmax'))
```

```
m1.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```
from keras.utils import to_categorical
```

```
m1.fit(dummy[:,1:65],to_categorical(dummy[:,65]),epochs=1,validation_split=0.5,verbose=2,shuffle=False)
```

```
m1.save_weights('/content/gdrive/My Drive/weightsMLround2')
```

```
test=np.array(pd.read_csv('test.csv'))
```

```
test[0]
```

```
res=m1.predict(test[:,1:])
```

```
res=np.argmax(res,axis=1)
```
